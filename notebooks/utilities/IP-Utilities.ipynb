# Creating animations with matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from shapely.geometry import Polygon
from shapely.affinity import translate
from IPython.display import HTML, display

# U-Kontur (einfaches Hufeisen)
u_coords = [
    (0, 0), (0, 40), (10, 40), (10, 10),
    (30, 10), (30, 40), (40, 40), (40, 0),
    (30, 0), (30, -10), (10, -10), (10, 0)
]
u_shape = Polygon(u_coords)

fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(0, 150)
ax.set_ylim(-20, 150)
ax.set_aspect("equal")
patch = plt.Polygon(u_shape.exterior.coords, closed=True, color="steelblue")

ax.add_patch(patch)
start = np.array([10, 10])
end = np.array([100, 100])
frames = 60

def update(frame):
    t = frame / (frames - 1)
    pos = (1 - t) * start + t * end
    moved = translate(u_shape, xoff=pos[0], yoff=pos[1])
    patch.set_xy(moved.exterior.coords)
    return patch,

anim = FuncAnimation(fig, update, frames=frames, interval=80, blit=True)
html = HTML(anim.to_jshtml())
display(html)
plt.close()
```
